```python
new_results.columns[0]
```

```python
new_results["% Ecotoxicity Freshwater (CTUe)"] = new_results["Ecotoxicity Freshwater (CTUe)"]/new_results["Ecotoxicity Freshwater (CTUe)"].sum()
new_results["% Ressources use (kg Sb eq.)"] = new_results["Ressources use (kg Sb eq.)"]/new_results["Ressources use (kg Sb eq.)"].sum()

new_results = new_results.drop(["Ecotoxicity Freshwater (CTUe)", "Ressources use (kg Sb eq.)"], axis = 1)
new_results = new_results.set_index("Types d'equipements")
```

```python
results = pd.read_csv("csv/Impacts_détaillé.csv", decimal=',')
```

```python
impact_target = ["BLD Utilisation des ressources, minéraux et métaux (kg Sb éq.) - ADPe", "BLD Ecotoxicité, eau douce (CTUe) - CTUe", "USE Utilisation des sols (sans dimensions) - LU"]

```

```python
import pandas as pd

# Assuming you have a DataFrame called 'df' and you want to rename a column

# Rename a single column
results.rename(columns={impact_target[0]: 'Ressource use - ADPe'}, inplace=True)
results.rename(columns={impact_target[1]: 'Ecotoxicity freshwater - CTUe'}, inplace=True)
results.rename(columns={impact_target[2]: 'Land use - LU'}, inplace=True)
```

```python
#targets = ['Basic electronic components', 'Sensors']
targets = ['Basic electronic components']

impact_target = ['Ressource use - ADPe', 'Ecotoxicity freshwater - CTUe', 'Land use - LU']
```

```python
goal_impact = impact_target[0]

goal_target = targets[0]
```

```python
target_results = results[results["Domaines"] == goal_target]

target_results.head()
target_results = target_results.set_index("Types d'équipements")
target_results = target_results.sort_values(by = goal_impact, ascending = False)
```

```python
target_results.head()
```

```python
pie = target_results[0:3].plot.pie(y=goal_impact, figsize=(5, 5), autopct='%1.1f%%')
pie.legend(loc='center', bbox_to_anchor=(1, 0.7))
plt.ylabel("")
plt.title('Contribution of elements inside block '+goal_target+' to '+goal_impact)
#plt.savefig('Contribution ADPe inside electronic comp')
plt.show()
```

```python
for t in targets:
    
    for im in impact_target:
        
        target_results = results[results["Domaines"] == t]
        target_results = target_results.set_index("Types d'équipements")
        target_results = target_results.sort_values(by = im, ascending = False)
        
        pie = target_results[0:3].plot.pie(y=goal_impact, figsize=(5, 5), autopct='%1.1f%%')
        pie.legend(loc='center', bbox_to_anchor=(0.5, 0))
        plt.ylabel("")
        plt.title(im)
        
        
        # Increase font size of autopct text
        for text in pie.texts:
            text.set_fontsize(20)
            
        for legend_text in pie.legend_.texts:
            legend_text.set_fontsize(17)
            
        plt.title(im, fontsize=20)
        #plt.savefig('GOOD Contribution of elements inside block '+t+' to '+im, bbox_inches='tight')
        
            
        plt.show()
        
        
    
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(len(impact_target), len(targets), figsize=(17, 17))
plt.subplots_adjust(wspace=0.5)  # Adjust the value of wspace to increase or decrease the space between columns

for i, t in enumerate(targets):
    for j, im in enumerate(impact_target):
        target_results = results[results["Domaines"] == t]
        target_results = target_results.set_index("Types d'équipements")
        target_results = target_results.sort_values(by=im, ascending=False)

        pie = target_results[0:3].plot.pie(y=goal_impact, autopct='%1.1f%%', ax=axes[j])
        pie.legend(loc='center', bbox_to_anchor=(0.5, 0))
        pie.set_ylabel("")
        pie.set_title(im, fontsize=17)

        # Increase font size of autopct text
        for text in pie.texts:
            text.set_fontsize(15)

        for legend_text in pie.legend_.texts:
            legend_text.set_fontsize(12)

plt.savefig('MAIN'+t+' to '+im, bbox_inches='tight')

plt.tight_layout()
plt.show()
```
