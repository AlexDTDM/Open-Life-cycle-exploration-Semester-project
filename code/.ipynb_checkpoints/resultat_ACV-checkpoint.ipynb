```python
to_drop = ['Resistances', "Connecteurs peripheriques", 'Capaciteurs', 'Conception', 'Others components', 'Transisteurs']



for i in results.index:

    if results["block"].loc[i] in to_drop: 
        results = results.drop(i)
        
```

```python
fig, axs = plt.subplots(4, 2, figsize=(25, 25))
j = 0
i = 0
for cat in results["indicator"].unique()[0:10]:
    
    
    
    test = results[results["indicator"] == cat].sort_values(by = 'value', ascending=False)
    test = test.set_index('block')
    
    if max(test["value"] > 0.001):
    
        #test.plot(kind='bar')
        axs[i,j].bar(test.index, test.value)
        axs[i,j].set_title(test["Detail"].unique(), fontsize=14, fontweight='bold')
        axs[i,j].set_ylabel(test["unit"].unique(), fontsize=14, fontweight='bold')
        axs[i, j].set_xticklabels([textwrap.fill(label, 15) for label in results["block"].unique()])
        #axs[i, j].tick_params(axis='x', labelsize=14)
    
        i = i + 1
        if i == 4:
            j = j + 1
            i = 0
plt.savefig('total impact part 1')        
plt.show()
```

```python
fig, axs = plt.subplots(4, 2, figsize=(25, 25))
j = 0
i = 0
for cat in results["indicator"].unique()[10:]:
    
    
    
    test = results[results["indicator"] == cat].sort_values(by = 'value', ascending=False)
    test = test.set_index('block')
    
    if max(test["value"] > 0.001):
    
        #test.plot(kind='bar')
        axs[i,j].bar(test.index, test.value)
        axs[i,j].set_title(test["Detail"].unique(), fontsize=14, fontweight='bold')
        axs[i,j].set_ylabel(test["unit"].unique(), fontsize=14, fontweight='bold')
        axs[i, j].set_xticklabels([textwrap.fill(label, 15) for label in results["block"].unique()])
        #axs[i, j].tick_params(axis='x', labelsize=14)
    
        i = i + 1
        if i == 4:
            j = j + 1
            i = 0
plt.savefig('total impact part 2')        
plt.show()
```

```python
contribution_indicateur = pd.read_csv("csv/contribution_des_indicateurs.csv")
contribution_indicateur = contribution_indicateur.set_index(["indicator"])
contribution_indicateur["%"] = contribution_indicateur["value"].apply(lambda a : (a/np.sum(contribution_indicateur["value"])*100))
sorted_contribution_indicateur = contribution_indicateur.sort_values(by="value", ascending = False)[0:]

sorted_contribution_indicateur
```

```python
sorted_contribution_indicateur.plot.pie(y="%", figsize=(5, 5), autopct='%1.1f%%')
plt.savefig('Contribution of indicators to the footprint')
plt.show()
```
